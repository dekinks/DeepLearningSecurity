```python
def plot_figure(F, A, p, model):
    l = 10

    fig, ax = plt.subplots()
    m = ax.bar(F, A, width = l)

    ax.set_xlim(10, 130)
    ax.set_ylim(90, 102)
    ax.set_xticks(F)

    ax.set_title('Accuracy of ' + str(model) + ' model with dynamic feature space', pad = 15)
    ax.set_ylabel('Accuracy (%)')
    ax.set_xlabel('Number of features')
    ax.bar_label(m, labels=[f'{x:.2%}' for x in p], size = 10, label_type = "edge")

    plt.savefig('AttackDetection_' + str(model) + '.png')
    plt.show()
```

```python
def plot_figure_classification(F, A, p, model):
    l = 0.3

    fig, ax = plt.subplots()
    m = ax.bar(F, A, width = l)

    ax.set_ylim(80, 102)
    ax.set_xlim(-0.5, 2.5)
    ax.set_xticks(F)

    ax.set_title('Accuracy of ' + str(model) + ' CNN model against KNN and NB', pad = 15)
    ax.set_ylabel('Accuracy (%)')
    ax.set_xlabel('Machine learning architecture')
    ax.bar_label(m, labels=[f'{x:.2%}' for x in p], size = 10, label_type = "edge")

    plt.savefig('AttackDetection_CNN_' + str(model) + '.png')
    plt.show()
```
