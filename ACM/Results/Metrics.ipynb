```python
bar_width = 0.3
x1 = np.arange(len(F_CNN))
x2 = [x + bar_width for x in x1]
x3 = [x + bar_width for x in x2]

print(range(len(F_CNN)))

plt.figure(figsize = (18, 9), dpi=200)

cnn = plt.bar(x1, F_CNN, width = bar_width, label = 'CNN', color = 'r')
knn = plt.bar(x2, F_KNN, width = bar_width, label = 'KNN', color = 'b')
nb = plt.bar(x3, F_NB, width = bar_width, label = 'NB', color = 'g')

plt.axis([-0.3, 10.0, 70.00, 105.00])
plt.xticks([x + bar_width for x in range(len(F_CNN))], Devices)

plt.title('Accuracy of each device according to the ML model')
plt.ylabel('Accuracy (%)')
plt.xlabel('Device')

ax = plt.gca()
plt.legend(bbox_to_anchor=(0, -0.15), bbox_transform=ax.transAxes, loc="lower left", ncol=len(F))

i = 0
for p in cnn:
  width = p.get_width()
  height = p.get_height()
  x, y = p.get_xy()

  plt.text(x + width / 2,
           y + height * 1.005,
           str(round(F_CNN[i], 1))+'%',
           ha='center',
           fontsize = 7)
  i += 1

j = 0
for p in knn:
  width = p.get_width()
  height = p.get_height()
  x, y = p.get_xy()

  plt.text(x + width / 2,
           y + height * 1.005,
           str(round(F_KNN[j], 1))+'%',
           ha='center',
           fontsize = 7)
  j += 1

k = 0
for p in nb:
  width = p.get_width()
  height = p.get_height()
  x, y = p.get_xy()

  plt.text(x + width / 2,
           y + height * 1.005,
           str(round(F_NB[k], 1))+'%',
           ha='center',
           fontsize = 7)
  k += 1

plt.show()
```
